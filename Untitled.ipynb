```python
# create an ampty "data" folder in the main folder before ruinning the code
import pandas as pd
epic_man_='https://webdata.illumina.com/downloads/productfiles/methylationEPIC/infinium-methylationepic-v-1-0-b5-manifest-file-csv.zip'
meth_='https://ftp.ncbi.nlm.nih.gov/geo/series/GSE142nnn/GSE142439/matrix/GSE142439_series_matrix.txt.gz'
```

```python
n=1
path='data/GSE142439_series_matrix.txt.gz'
df=pd.read_csv(path,sep='\t',skiprows=71,index_col=0) # methylation data starts from line 71
df.shape
```

```python
df=df.drop('!series_matrix_table_end')
meta=pd.read_csv(path,sep='\t',skiprows=40,index_col=0,nrows=3) # meta data starts from line 40
meta.columns=df.columns
meta.index=['age','cell','treatment']
meta=meta.T
meta['age']=meta.age.str.strip('age: ')
meta['cell']=meta.cell.str.strip('cell type: ')
meta['treatment']=meta.treatment.str.strip('treatmentstatus: ')
meta['t']=meta.treatment.apply(lambda x: 1 if x=='Normal' else 0)
meta['cell']=meta.cell.apply(lambda x: 'skin' if x=='skin fibroblasts' else 'vein')
meta.T
```

```python
for tissue in ['skin','vein']:
    df[tissue+'_normal']=df[meta[(meta.cell==tissue)&(meta.treatment=='Normal')].index].mean(1)
    df[tissue+'_treated']=df[meta[(meta.cell==tissue)&(meta.treatment=='Treated')].index].mean(1)
    df[tissue+'_dif']=df[tissue+'_treated']-df[tissue+'_normal']
df=df.drop(meta.index,axis=1)
df.head(1)
```

```python
man='infinium-methylationepic-v-1-0-b5-manifest-file-csv.zip'
man=pd.read_csv('data/'+man,skiprows=7,index_col=0)
man.head(1)
```

```python
df=man[['CHR_hg38','Start_hg38','End_hg38']].merge(df,left_index=True,right_index=True)
df.shape
```
